```python
import time
import unittest
```

```python
# Reuse Existing Functions
def calculate_total_price(price_per_unit, quantity):
    return price_per_unit * quantity

def check_availability(available, requested):
    return requested <= available

def apply_discount(total_price, discount_percent):
    return total_price * (1 - discount_percent / 100)
```

```python
# Integrated Workflow for Performance
def book_service(price_per_unit, quantity, available, discount_percent):
    if not check_availability(available, quantity):
        return "Not Available"
    
    total = calculate_total_price(price_per_unit, quantity)
    discounted_total = apply_discount(total, discount_percent)
    return discounted_total
```

```python
#Single Operation Performance Test
# Measure execution time for a single booking
start_time = time.time()

result = book_service(100, 2, 5, 10)

end_time = time.time()
elapsed_time = end_time - start_time

print(f"Booking result: {result}")
print(f"Execution time: {elapsed_time:.6f} seconds")
```

```python
#Load Testing â€“ Multiple Bookings
# Simulate 1000 bookings
num_requests = 1000
start_time = time.time()

for _ in range(num_requests):
    book_service(100, 2, 5, 10)

end_time = time.time()
total_time = end_time - start_time
avg_time_per_request = total_time / num_requests

print(f"Total time for {num_requests} bookings: {total_time:.6f} seconds")
print(f"Average time per booking: {avg_time_per_request:.6f} seconds")
```
